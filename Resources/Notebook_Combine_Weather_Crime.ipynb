```python
import pandas as pd
from datetime import datetime

# import crime data
Crime_data = pd.read_csv('Crime_data_2016.csv')
Crime_data = Crime_data[['Case Number','Primary Type','Date','DateConvert']]
Crime_data.head()
```

```python
#import weather data
Weather_data =pd.read_csv('temperature_2016.csv')
Weather_data.head()
```

```python
#merge 2 datasets
Crime_Weather = Crime_data.merge(Weather_data, on='DateConvert')

Crime_Weather.head()
```

```python
# select columns that we will use
Crime_Weather = Crime_Weather[['Case Number','Primary Type','datetime','Temp_Farenheit']]

print(Crime_Weather.head())
print()

Crime_Weather.info()
```

```python
# export data to a csv file 
Crime_Weather.to_csv ('Crime_Weather_2016.csv',encoding='utf-8')
```
